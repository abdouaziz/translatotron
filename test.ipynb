```python
import torch 
import torch.nn as nn 
from translatotron import Translatotron , play_audio
```

```python
 
model = Translatotron()
input_mel = torch.randn(1, 100, 80)  # Example input

# Generate waveform
waveform, _, _ = model(input_mel)

# Play or save the audio
audio_output = play_audio(waveform[0])  # Use the first item if it's a batch

# If in a notebook, display the audio widget
if audio_output is not None:
    display(audio_output)

# Optionally, save to a file
play_audio(waveform[0], filename="translated_speech.wav")
```
