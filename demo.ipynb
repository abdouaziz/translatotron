```python
import torch 
from translatotron import Translatotron , play_audio
```

```python
 
model = Translatotron()
input_mel = torch.randn(1, 100, 80)  # Example input

# Generate waveform
waveform, _, _ , _= model(input_mel)

# Play or save the audio
audio_output = play_audio(waveform[0])  # Use the first item if it's a batch

# If in a notebook, display the audio widget
if audio_output is not None:
    display(audio_output)

# Optionally, save to a file
play_audio(waveform[0], filename="translated_speech.wav")
```

```python

x = torch.randn(1, 100, 80)  # Batch size 1, 100 time steps, 80 features (log-mel spectrogram)

# Test with speaker reference
speaker_reference = torch.randn(
    1, 80
)  # Batch size 1, 80 features (speaker reference utterance)
waveform, spectrogram, phonemes_en, phonemes_es = model(x , speaker_reference=speaker_reference)
print("With speaker reference:")
print(f"Waveform shape: {waveform.shape}")
print(f"Spectrogram shape: {spectrogram.shape}")
print(f"English phonemes shape: {phonemes_en.shape}")
print(f"Spanish phonemes shape: {phonemes_es.shape}")

# Test without speaker reference
waveform, spectrogram, phonemes_en, phonemes_es = model(x)
print("\nWithout speaker reference:")
print(f"Waveform shape: {waveform.shape}")
print(f"Spectrogram shape: {spectrogram.shape}")
print(f"English phonemes shape: {phonemes_en.shape}")
print(f"Spanish phonemes shape: {phonemes_es.shape}")
```
